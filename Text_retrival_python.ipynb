# 1 Import các thưu viện cần thiết

```python
import string
import nltk
nltk.download('stopwords')
nltk.download('punkt')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
```

# 2 đọc bộ dữ liệu

```python
DATA_PATH = '/content/2cls_spam_text_cls.csv'
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
messages = df['Message'].tolist()
labels = df['Category'].tolist()
```

```python
def transform_labels(labels):
    return [1 if label == 'spam' else 0 for label in labels]
y = transform_labels(labels)
print(f'Encoded labels: {y}')
```

# 4 tiền xử lí dữ liệu

```python
def lowercase(text):
    return text.lower()

def punctuation_removal(text):
    translator = str.maketrans('', '', string.punctuation)

    return text.translate(translator)

def tokenize(text):
    return nltk.word_tokenize(text)

def remove_stopwords(tokens):
    stop_words = nltk.corpus.stopwords.words('english')

    return [token for token in tokens if token not in stop_words]

def stemming(tokens):
    stemmer = nltk.PorterStemmer()

    return [stemmer.stem(token) for token in tokens]

def preprocess_text(text):
    text = lowercase(text)
    text = punctuation_removal(text)
    tokens = tokenize(text)
    tokens = remove_stopwords(tokens)
    tokens = stemming(tokens)

    return tokens
```

```python
messages = [preprocess_text(message) for message in messages]
```

```python
def create_dictionary(messages):
    dictionary = []

    for tokens in messages:
        for token in tokens:
            if token not in dictionary:
                dictionary.append(token)

    return dictionary

def create_features(tokens, dictionary):
    features = np.zeros(len(dictionary))

    for token in tokens:
        if token in dictionary:
            features[dictionary.index(token)] += 1

    return features
```

```python
dictionary = create_dictionary(messages)
X = np.array([create_features(tokens, dictionary) for tokens in messages])
print(f'Dictionary: {len(dictionary)}')
print(f'Shape of X: {X.shape}')
```

# 5. Chia bộ dữ liệu train/val/test:

```python
VAL_SIZE = 0.2
TEST_SIZE = 0.125
SEED = 0

X_train, X_val, y_train, y_val = train_test_split(X, y,
                                                  test_size=VAL_SIZE,
                                                  shuffle=True,
                                                  random_state=SEED)
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train,
                                                    test_size=TEST_SIZE,
                                                    shuffle=True,
                                                    random_state=SEED)
```

# 6. Huấn luyện mô hình:

```python
class NaiveBayes:
    def __init__(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train

    def get_labels(self):
        unique_labels = set(self.y_train)
        sorted_unique_labels = sorted(unique_labels)
        return sorted_unique_labels

    def get_index_by_label(self):
        label = self.get_labels()
        index_labels = []
        for i in range(len(label)):
            index_label = [index for index, value in enumerate(self.y_train) if value == label[i]]
            index_labels.append(index_label)
        return index_labels

    def laplacian_smoothing(self, array):
        zero_count = np.sum(array == 0)
        array[array == 0] = 1
        return array

    def fit(self):
        labels_count = []
        index_labels = self.get_index_by_label()

        for index_label in index_labels:
            label_count = np.sum(self.X_train[index_label, :], axis=0)
            labels_count.append(label_count)

        for i in range(len(labels_count)):
            labels_count[i] = self.laplacian_smoothing(labels_count[i])
            labels_count[i] = labels_count[i] / np.sum(labels_count[i])

        return labels_count

    def messages_features(self, messages, dictionary):
        features = np.zeros(len(dictionary))

        for token in messages:
            if token in dictionary:
                features[dictionary.index(token)] += 1

        return features

    def predict(self, object):
        labels_count = self.fit()
        get_labels = self.get_labels()
        get_predict = []
        # Thay thế giá trị 0 bằng NaN để không ảnh hưởng đến tính toán tổng
        for i in range(len(labels_count)):
            labels_count_predict = np.where(object[i] > 0, labels_count[i]*np.log(labels_count[i]), 0)
            labels_count_predict = np.sum(labels_count_predict)
            get_predict.append(labels_count_predict)

        index_max = np.argmax(get_predict)
        return get_labels[index_max]
```

```python
test_input = 'I am actually thinking a way of doing something useful'
model = NaiveBayes(X_train, y_train)
# model.predict()

y_test_pred = []
for i in range(len(X_test)):
    y_test_pred.append(model.predict(X_test[i]))

y_val_pred = []
for i in range(len(X_val)):
    y_val_pred.append(model.predict(X_val[i]))
```

```python
def accuracy_score(y_true, y_pred):
    count = sum(1 if y_true[i] == y_pred[i] else 0 for i in range(len(y_true)))

    return count / len(y_true)

accuracy_score_test = accuracy_score(y_test, y_test_pred)
accuracy_score_val = accuracy_score(y_val, y_val_pred)
print(f'Accuracy score test: {accuracy_score_test}')
print(f'Accuracy score val: {accuracy_score_val}')
```
